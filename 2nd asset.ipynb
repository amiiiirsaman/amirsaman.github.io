```python
import pandas as pd
import matplotlib.pyplot as plt
corona_economy = pd.read_csv('Section 1.csv')
```

```python
import datetime

FL_covid_new_case = corona_economy['FL_new_case_count']
CA_covid_new_case = corona_economy['CA_new_case_count']
TX_covid_new_case = corona_economy['TX_new_case_count']

FL_spending = corona_economy['FL_spend_all']
CA_spending = corona_economy['CA_spend_all']
TX_spending = corona_economy['TX_spend_all']

FL_revenue = corona_economy['FL_revenue_all']
CA_revenue = corona_economy['CA_revenue_all']
TX_revenue = corona_economy['TX_revenue_all']

FL_merchant = corona_economy['FL_merchants_all']
CA_merchant = corona_economy['CA_merchants_all']
TX_merchant = corona_economy['TX_merchants_all']

FL_employment = corona_economy['FL_emp']
CA_employment = corona_economy['CA_emp']
TX_employment = corona_economy['TX_emp']

FL_spending_apparel_accessories = corona_economy['FL_spend_aap']
FL_spending_accomodation_food_service = corona_economy['FL_spend_acf']
FL_spending_entertainment_recreation = corona_economy['FL_spend_aer']
FL_spending_grocery_food = corona_economy['FL_spend_grf']
FL_spending_general_merchandise = corona_economy['FL_spend_gen']
FL_spending_home_improvement = corona_economy['FL_spend_hic']
FL_spending_health_care = corona_economy['FL_spend_hcs']
FL_spending_sporting_goods = corona_economy['FL_spend_sgh']
FL_spending_transportation = corona_economy['FL_spend_tws']

FL = corona_economy[['FL_new_case_count', 'FL_spend_all', 'FL_revenue_all', 'FL_merchants_all', 'FL_emp','FL_spend_aap', 'FL_spend_acf','FL_spend_aer', 'FL_spend_grf', 'FL_spend_gen', 'FL_spend_hic', 'FL_spend_hcs', 'FL_spend_sgh']]
```

```python
plt.hist(FL['FL_new_case_count'])
plt.title('Distribution of FL_new_case_count')
plt.xlabel('FL New Covid 19 Cases')
plt.show()
```

```python
plt.scatter(FL['FL_new_case_count'],
            FL['FL_revenue_all'])

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

Covid_High = FL[FL['FL_new_case_count'] >= -.4]
frequencies = Covid_High.sum()
frequencies.plot.barh()
plt.show()
```

```python
FL_spending_types = ['FL_spend_aap', 'FL_spend_acf','FL_spend_aer', 'FL_spend_grf', 'FL_spend_gen', 'FL_spend_hic', 'FL_spend_hcs', 'FL_spend_sgh']
plt.figure(figsize=(12, 6))
   
for x in FL_spending_types:
    plt.plot(corona_economy['Time'], FL[x], label=x)

plt.legend()
plt.show()
```
